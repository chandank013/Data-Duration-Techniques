```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the IPL auction page
url = "https://www.iplt20.com/auction/2022"

# Send a GET request to the URL and store the response
response = requests.get(url)

# Parse the content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the auction table in the parsed HTML
table = soup.find("table", class_="ih-td-tab auction-tbl")

# Extract the header row from the table
title = table.find_all("th")

# Initialize a list to store the header names
header = []
for i in title:
    name = i.text  # Get the text of each header element
    header.append(name)  # Append the header name to the list

# Create a DataFrame with the extracted header names
df = pd.DataFrame(columns=header)

# Extract all rows from the table
rows = table.find_all("tr")

# Loop through each row (skipping the header row) to extract cell data
for i in rows[1:]:
    
    # Find the first cell which contains a specific div class
    first_td = i.find_all("td")[0].find("div", class_="ih-pt-ic")
    # Extract all subsequent cells in the row
    data = i.find_all("td")[1:]
    # Create a list of text values from the data cells
    row = [tr.text for tr in data]
    # Insert the first cell's text at the beginning of the row list
    row.insert(0, first_td)
    # Get the current length of the DataFrame to append the new row
    l = len(df)
    # Append the new row to the DataFrame
    df.loc[l] = row
    
# print(df)  # Uncomment to see the populated DataFrame

# Export the DataFrame to a CSV file
df.to_csv("ipl auction stats.csv")

# Print a confirmation message indicating the data export
print("Data has been exported to ipl auction stats.csv.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

url = "https://www.espncricinfo.com/records/tournament/batting-most-runs-career/indian-premier-league-2016-11001"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table", {"class": 'ds-table'})

# Extract rows from the table
rows = table.find_all("tr")

# Open a file to write data
with open('ipl_2016_batting_records.csv', 'w', newline='') as file:
    writer = csv.writer(file)

    # Loop through rows and extract cell data
    for row in rows:
        cells = row.find_all("td")  # Use "th" if header cells
        cell_data = [cell.text.strip() for cell in cells]
        
        # Write the row to the CSV file
        writer.writerow(cell_data)

print("Data has been exported to ipl_2016_batting_records.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

url = "https://en.m.wikipedia.org/wiki/List_of_tallest_voluntarily_demolished_buildings"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table", {"class": 'wikitable sortable'})

# Extract rows from the table
rows = table.find_all("tr")

# Open a file to write data
with open('wiki_build.csv', 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)

    # Loop through rows and extract cell data
    for row in rows:
        cells = row.find_all("td")  # Use "th" if header cells
        cell_data = [cell.text.strip() for cell in cells]
        
        # Write the row to the CSV file
        writer.writerow(cell_data)

print("Data has been exported successfully")
```
